```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=100
std=0.5
balloons_width=np.random.normal(8,std,N)
balloons_length=np.random.normal(8,std,N)
balloons_weight=np.random.normal(12,std,N)
```

```python
melon_width=np.random.normal(6,std,N)
melon_length=np.random.normal(12,std,N)
melon_weight=np.random.normal(15,std,N)
```

```python
width=np.concatenate((balloons_width,melon_width))
length=np.concatenate((balloons_length,melon_length))
weight=np.concatenate((balloons_weight,melon_weight))
```

```python
x=np.array([width,length,weight]).T
```

```python
melon_label=np.zeros(N,dtype='int')
balloons_label=np.ones(N,dtype='int')
y=np.concatenate((melon_label,balloons_label))
```

```python
from mpl_toolkits.mplot3d import Axes3D
p=plt.figure()
fig=Axes3D(p)
fig.scatter(x[0:N,0],x[0:N,1],x[0:N,2], c='blue')
fig.scatter(x[N:,0],x[N:,1],x[N:,2], c='pink')

fig.set_xlabel('widths')
fig.set_ylabel('lengths')
fig.set_zlabel('weights')
plt.show()
```
